```python
from dash import Dash, html, dcc, callback, Output, Input
import plotly.express as px
import pandas as pd

df = pd.read_csv(
    "https://raw.githubusercontent.com/plotly/datasets/master/gapminder_unfiltered.csv"
)

app = Dash(__name__)

app.layout = [
    html.H1(
        children="Title of Dash App",
        style={"textAlign": "center"}
    ),
    dcc.Dropdown(
        df.country.unique(),
        "Canada",
        id="dropdown-selection"
    ),
    dcc.Graph(id="graph-content")
]

@callback(
    Output("graph-content", "figure"),
    Input("dropdown-selection", "value")
)
def update_graph(value):
    dff = df[df.country == value]
    return px.line(dff, x="year", y="pop")

if __name__ == "__main__":
    app.run(debug=True)
```

```python
import pandas as pd
import numpy as np

from dash import Dash, dcc, html, Input, Output
import plotly.express as px
from sklearn.cluster import KMeans

# 1. Preparar datos
np.random.seed(42)

df = pd.DataFrame(
    np.random.randn(200, 2),
    columns=["X", "Y"]
)

# 2. Crear la app
app = Dash(__name__)

app.layout = html.Div([
    html.H2("Visualizador de Clustering K-Means"),
    html.P("Mueve el slider para cambiar el número de grupos"),
    
    dcc.Slider(
        id="k-slider",
        min=2,
        max=10,
        step=1,
        value=3,
        marks={i: str(i) for i in range(2, 11)}
    ),
    
    dcc.Graph(id="cluster-graph")
])

# 3. Callback
@app.callback(
    Output("cluster-graph", "figure"),
    Input("k-slider", "value")
)
def update_graph(k_clusters):
    model = KMeans(
        n_clusters=k_clusters,
        init="k-means++",
        n_init=10,
        random_state=42
    )

    df["cluster"] = model.fit_predict(df[["X", "Y"]]).astype(str)

    fig = px.scatter(
        df,
        x="X",
        y="Y",
        color="cluster",
        title=f"K-Means con {k_clusters} grupos"
    )

    return fig

# 4. Ejecutar en Jupyter Notebook local
if __name__ == "__main__":
    app.run(debug=True, port=8051)
```

```python
from dash import Dash, html, dcc, callback, Output, Input
import plotly.express as px
import pandas as pd

# Cargar datos
df = pd.read_csv(
    "https://raw.githubusercontent.com/plotly/datasets/master/gapminder_unfiltered.csv"
)

# Crear aplicación
app = Dash(__name__)

# Layout
app.layout = [
    html.H1(
        children="Title of Dash App",
        style={"textAlign": "center"}
    ),

    dcc.Dropdown(
        options=df.country.unique(),
        value="Canada",
        id="dropdown-selection"
    ),

    dcc.Graph(id="graph-content")
]

# Callback
@callback(
    Output("graph-content", "figure"),
    Input("dropdown-selection", "value")
)
def update_graph(value):

    dff = df[df["country"] == value]

    fig = px.line(
        dff,
        x="year",
        y="pop",
        title=f"Population of {value}"
    )

    return fig

# Ejecutar aplicación
if __name__ == "__main__":
    app.run(debug=True, port=8050)
```

```python
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd

# Cargar datos
df = pd.read_csv(
    "https://raw.githubusercontent.com/plotly/datasets/master/gapminder_unfiltered.csv"
)

# Crear gráfico con todos los países
fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    size="lifeExp",
    color="continent",
    hover_name="country",
    animation_frame="year",
    size_max=55,
    log_x=True,
    title="Relación entre PIB per cápita y esperanza de vida"
)

# Crear app
app = Dash(__name__)

app.layout = [
    html.H1(
        "Gapminder - Todos los países",
        style={"textAlign": "center"}
    ),
    dcc.Graph(
        id="graph-content",
        figure=fig
    )
]

# Ejecutar en Jupyter
app.run(debug=True, port=8050)
```
